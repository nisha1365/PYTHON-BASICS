```python
class human:
  def __init__(this):
    this.name = "stefan doe"
    this.age = 34
    this.currency = 'USD'
    this.hungerStatus = True
  def setNewProperties(this, name, age):
    this.name = name
    this.age = age
  def getHumanProperties(this):
    return this.name, this.age, this.currency, ("hungry" if this.hungerStatus==True else "Not Hungry")
```

```python
info = human()
info.getHumanProperties()
```

```python
class abc:
  def mul(this, a,b):
    return a*b
  def mul(this, a):
    return a*a
```

```python
class ArithmeticProg:
  def __iter__(self):
    self.base = -3
    self.step = 2
    return self
  def __next__(self):
    self.base = self.base + self.step
    return self.base
```

```python
obj = ArithmeticProg()
i = iter(obj)
```

```python
for j in range(14):
  print(next(i))
```
