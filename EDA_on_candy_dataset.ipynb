```python
import pandas as pd
```

```python
df = pd.read_csv('/content/candy.csv')
```

```python
df.head()
```

```python
# importing the required module
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

1. SWARMPLOT

```python
sns.swarmplot(x='caramel', y='pricepercent', data=df)
```

2. STRIPPLOT

```python
sns.stripplot(x='chocolate',y='pricepercent',data=df)
```

3. BARPLOT

```python
sns.barplot(x ='chocolate', y ='pricepercent', data = df,  
            palette ='plasma')
```

4. COUNTPLOT

```python
sns.countplot(x ='fruity', data = df) 
```

5.BOXPLOT

```python
sns.boxplot(x='fruity', y='pricepercent', data=df, hue='chocolate')
```

6. VIOLINPLOT

```python
sns.violinplot(x='fruity',y='sugarpercent',data=df)
```

7. PAIRPLOT

```python
sns.pairplot(df)
```

8.SCATTER PLOT

```python
sns.scatterplot(x='nougat',y= 'winpercent',data = df)
```

9.DISTPLOT

```python
sns.distplot(df['winpercent'])
```

10.JOINTPLOT

```python
sns.jointplot(x='nougat',y='sugarpercent',data=df)
```
